```python
import matplotlib.pyplot as plt

# Data for two datasets
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]  # Dataset 1
y2 = [1, 3, 5, 7, 9]   # Dataset 2

# Plotting the datasets
plt.plot(x, y1, label='Dataset 1', color='blue', linestyle='--')  # Add label for Dataset 1
plt.plot(x, y2, label='Dataset 2', color='red', linestyle='-')   # Add label for Dataset 2

# Adding title and labels
plt.title("Line Plot with Legend")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Adding the legend
plt.legend(loc=5)

# Display the plot
plt.show()
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
x
```

```python
#functional 
plt.plot(x,y, marker='*')
#plt.show() show the plot
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
#SubPloting
plt.subplot(1,2,1)
plt.plot(x,y,'-r')

plt.subplot(1,2,2)
plt.plot(y,x,'-y')
```

```python
#OO
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
# for current_ax in axes:
#     current_ax.plot(x,y)
axes[0].plot(x,y)
axes[1,].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout() #no overlaping
```

Figure Size and Dpi

```python
fig =plt.figure(figsize=(8,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(figsize=(8,2))
axes.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y,color='red')
axes[1].plot(x,y)
plt.tight_layout()
```

Plot Appearance

```python
fig=plt.figure()
ax=fig.add_axes([0,0,0.5,0.5])
ax.plot(x,y,color='red',linewidth=1,ls="-",marker='o',markersize=10,markerfacecolor='yellow',markeredgewidth=2,markeredgecolor="blue")
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,0.5,0.5])
ax.plot(x,y,ls="--")
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
