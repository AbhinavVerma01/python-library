Bar Chart in 

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array(["A","B","C","D","E"])
y=np.array([12,3,4.5,5.6,6])
plt.bar(x,y)
```

Horizontal Bars:
If you want the bars to be displayed horizontally instead of vertically, use the barh() function

```python
x=np.array(["A","B","C","D","E"])
y=np.array([12,3,4.5,5.6,6])
plt.barh(x,y)
```
