Bar Chart in 

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array(["A","B","C","D","E"])
y=np.array([12,3,4.5,5.6,6])
plt.bar(x,y,width=0.4)
```

Horizontal Bars:
If you want the bars to be displayed horizontally instead of vertically, use the barh() function

```python
x=np.array(["A","B","C","D","E"])
y=np.array([12,3,4.5,5.6,6])
plt.barh(x,y,color="hotpink", height = 0.3)
```

```python
# Histogram
# A histogram is a graph showing frequency distributions.

# It is a graph showing the number of observations within each given interval.
```

```python
x=np.random.normal(170, 10, 250)
plt.hist(x)
plt.show()
```

```python
# Pie Charts
x=np.array([70,20,40,50])
lab=["Apple","Banana","Guava","Dates"]
plt.pie(x,labels=lab,startangle =90)
plt.show()
```

```python
x=np.array([70,20,40,50])
lab=["Apple","Banana","Guava","Dates"]
exp=[0.2,0,0,0]
plt.pie(x,labels=lab,explode=exp,shadow=True)
plt.show()
```

```python
x=np.array([70,20,40,50])
lab=["Apple","Banana","Guava","Dates"]
plt.pie(x,labels=lab,)
plt.legend(loc=2,title="For Fruits :")
plt.show()
```
