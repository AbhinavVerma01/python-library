```python
import numpy as np
x=np.arange(0,10)
y=x*2
z=x**2
```

```python
print(x)
print(y)
print(z)
```

```python
import matplotlib.pyplot as plt
```

** Create a figure object called fig using plt.figure() **
** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig =plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Title')


```

Exercise 3Â¶
** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**
** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**


```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.4,.4])

ax1.plot(x,z)
ax1.set_xlabel('X Label')
ax1.set_ylabel('Z Label')


ax2.plot(x,y)
ax2.set_xlabel('X Label')
ax2.set_ylabel('Y Label')
ax2.set_title('Inner')
ax2.set_xlim(20,22)
ax2.set_ylim(30,55)
```

Exercise 4Â¶
** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**
** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
fig,axes=plt.subplots(ncols=2,nrows=1)
axes[0].plot(x,y,color='red',lw="2",ls="--")
axes[1].plot(x,z,color='blue',lw="2",ls="--")
```

```python
fig,axes=plt.subplots(figsize=(14,2),ncols=2,nrows=1)
axes[0].plot(x,y,color='red',lw="2",ls="--")
axes[0].set_xlabel('X Label')
axes[0].set_ylabel('Y Label')

axes[1].plot(x,z,color='blue',lw="2",ls="--")
axes[1].set_xlabel('X Label')
axes[1].set_ylabel('Z Label')
axes[1].grid(color = 'green', linestyle = '-', linewidth = 0.5)
```
