```python
df = df[~df['Short name'].str.startswith('@',na=False)]
df.head()
```

```python
df["site_id"] = df["Short name"].apply(
    lambda x: x.split("_")[4][:-1] if "_" in x and len(x.split("_")) > 4
    else x.split("-")[3][:-1] if "-" in x and len(x.split("-")) > 3
    else x[:-1])
df.head()
```

```python
df["sectors"] = df["Short name"].apply(
    lambda x: x.split("_")[4] if "_" in x and len(x.split("_")) > 4
    else x.split("-")[3] if "-" in x and len(x.split("-")) > 3
    else None)
df.head()
```
