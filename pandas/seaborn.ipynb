Transform()

```python
df=pd.DataFrame({
    "Group":["x","y","z","x","y","z"],
    "val1":[2,5,10,20,42,32],
    "val2":[23,54,19,209,424,324]
})
df
```

x'=x-min(x)/max(x)-min(x)

```python
df_new=df.loc[:,"val1":"val2"]
df_new
```

```python
def normalize(x):
    return (x-x.min())/(x.max()-x.min())
```

```python
df_new.transform(normalize)
```

```python
df_new.loc[1,"val2"]
```

```python
df_new.transform(lambda a:np.sin(a))
```

```python
np.sin(df_new.val2)
```
