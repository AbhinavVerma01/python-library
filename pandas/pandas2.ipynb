Multi Index And Index Hierarchy



```python
import numpy as np
import pandas as pd
```

```python
inside=["ClassA","ClassB","ClassC","ClassA","ClassB","ClassC"]
outside=["School1","School1","School1","School2","School2","School2"]
zip(outside,inside)
```

```python
multi_index=list(zip(outside,inside))
multi_index
```

```python
hier_index=pd.MultiIndex.from_tuples(multi_index)
hier_index
```

```python
np.random.seed(101)
df=pd.DataFrame(np.random.randint(70,100,size=(6,2)),index=hier_index,columns=["1st_Semester","2ed_semester"])
df
```

Concatentnating Pandas Dataframe:Concat() Function

```python
df1=pd.DataFrame({"X":["x0","x1","x2","x3"],
                  "Y":["y0","y1","y2","y3"],
                  "Z":["z0","z1","z2","z3"],
                  "XL":["xl0","xl1","xl2","xl3"]})

df2=pd.DataFrame({"X":["x4","x5","x6","x7"],
                  "Y":["y4","y5","y6","y7"],
                  "Z":["z4","z5","z6","z7"],
                  "XL":["xl4","xl5","xl6","xl7"]})

df3=pd.DataFrame({"X":["x8","x9","x10","x11"],
                  "Y":["y8","y9","y10","y11"],
                  "Z":["z8","z9","z10","z11"],
                  "XL":["xl8","xl9","xl10","xl11"]})
df1
```

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3],ignore_index=True)
```

```python
df1.columns
```

```python
df2.columns=["X","Y","Z","A"]
```

```python
df2
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.concat([df1,df2],ignore_index=True,join='inner')
```

```python
pd.concat([df1,df2,df3],axis=1)
```

```python
df1=pd.DataFrame({"X":["x0","x1","x2","x3"],
                  "Y":["y0","y1","y2","y3"],
                  "Z":["z0","z1","z2","z3"],
                  "XL":["xl0","xl1","xl2","xl3"]},
                 index=[1,2,3 ,4])

df2=pd.DataFrame({"X":["x4","x5","x6","x7"],
                  "Y":["y4","y5","y6","y7"],
                  "Z":["z4","z5","z6","z7"],
                  "XL":["xl4","xl5","xl6","xl7"]},
                  index=[5,6,7,8])

df3=pd.DataFrame({"X":["x8","x9","x10","x11"],
                  "Y":["y8","y9","y10","y11"],
                  "Z":["z8","z9","z10","z11"],
                  "XL":["xl8","xl9","xl10","xl11"]},
                  index=[9,10,11,12])
df1
```

```python
pd.concat([df1,df2,df3],axis=1)
```

```python
pd.concat([df1,df2,df3],axis=1,join='inner')
```

Merge() in Data Frame

```python
df1=pd.DataFrame({"Employee":["Max","John","Adam","Nik"],
                  "Department":["DS","WD","CS","DS"],
                  "Year":["2003","2008","2023","2024"]})
df2=pd.DataFrame({"Employee":["Harry","Adam","Nik","Max"],
                  "State":["UP","UK","MP","HR"],
                  "Slary":["20000","28000","23900","20240"]})
df1
```

```python
df2
```

```python
pd.concat([df1,df2],axis=1 )
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on="Employee")
```

```python
df3=pd.DataFrame({
    "Manager":["Abram","Joseph","Kayra"],
    "Department":["DS","WD","CS"]
})
df3
```

```python
df4=pd.merge(df1,df2)
df4
```

```python
pd.merge(df4,df3,on="Department")
```

```python
df3=pd.merge(df1,df2)
df3
```

```python
df5=pd.DataFrame({
    "Member":["Max","Nik","John"],
    "Family_member":["1","4","6"]
})
df5
```

```python
pd.merge(df3,df5,left_on="Employee",right_on="Member")
```

```python
pd.merge(df3,df5,left_on="Employee",right_on="Member",how="outer")
```

JOIN()::

```python
df1=pd.DataFrame({
    "X":["X0","X1","X3"],
    "Y":["Y1","Y2","Y3"]
},index=["A0","A1","A2"])

df2=pd.DataFrame({
    "Z":["Z0","Z1","Z3"],
    "T":["T1","T2","T3"]
},index=["A0","A2","A3"])
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2,how="inner")
```

```python
df1.join(df2,how="outer")
```

```python
df1.join(df2,how="right")
```

```python
df1.join(df2,how="left")
```

```python
df3.join(df4,lsuffix="left",rsuffix="right")
```

df3.join(df4,lsuffix="left",rsuffix="right")

```python
df3.join(df4,lsuffix="left",rsuffix="right",how="inner")
```

```python
df3.join(df4,lsuffix="left",rsuffix="right",how="outer")
```
