```python
import matplotlib.pyplot as plt

def lineplot(x, y, title, xlabel, ylabel):
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
lineplot(x, y, 'Line Plot Example', 'X Axis', 'Y Axis')
```

```python
# Displot with faceting
sns.displot(data=tips, x="total_bill", hue="sex", kind="kde")
plt.show()
```

```python
# Boxplot
sns.boxplot(data=tips, x="total_bill", y="day", hue="smoker", palette='coolwarm')
# Common Seaborn Palettes:
# Qualitative Palettes (best for categorical data):
# deep, muted, bright, pastel, dark, colorblind, Set1, Set2
# Diverging Palettes (best for displaying diverging data, like temperature):
# coolwarm, RdBu, PuOr, BrBG, PiYG
# Sequential Palettes (best for ordered data like heatmaps):
# Blues, Greens, Purples, Oranges
plt.show()
```

```python
# Violinplot
sns.violinplot(data=tips, x="total_bill", y="day", hue="sex", palette="coolwarm")
plt.show()
#When to Use: When you want to show both summary statistics and the shape of the data distribution
# for each category
```

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Dummy Weather Data
data = {
    "Temperature": [25, 30, 35, 28, 24, 32],
    "Humidity": [60, 55, 45, 70, 80, 50],
    "Wind Speed": [12, 14, 20, 8, 6, 18],
    "Rainfall": [0, 5, 10, 2, 0, 7]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Features (X) and Target (y)
X = df[["Temperature", "Humidity", "Wind Speed"]]
y = df["Rainfall"]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("R-squared (R²):", r2)

# Display the model coefficients
print("Coefficients (Temperature, Humidity, Wind Speed):", model.coef_)
print("Intercept:", model.intercept_)

# Visualize Actual vs Predicted Rainfall
plt.scatter(range(len(y_test)), y_test, color="blue", label="Actual")
plt.scatter(range(len(y_test)), y_pred, color="red", label="Predicted")
plt.xlabel("Sample Index")
plt.ylabel("Rainfall (mm)")
plt.title("Actual vs Predicted Rainfall")
plt.legend()
plt.show()
```
