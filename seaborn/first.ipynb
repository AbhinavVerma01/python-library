Distribution Plots


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Distribution Plots
tips=sns.load_dataset('tips')
tips.head()
```

```python
# Create a distribution plot for the 'total_bill' column
sns.displot(tips['total_bill'] ,kde=False ,bins=60)

# Show the plot
plt.grid()
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips) #kind='hex'
plt.grid()
plt.show()
```

```python
sns.pairplot(tips,hue='sex')
plt.show()
```

```python
sns.rugplot(tips['total_bill'])
plt.show()
```

```python
# Categorical Plots
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,palette="viridis")
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
sns.swarmplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',)
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
#Matrix plot
tips=sns.load_dataset('tips')
flight=sns.load_dataset('flights')
tips.head()
```

```python
flight.head()
```

```python
tc = tips.select_dtypes(include='number').corr()
tc

```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
plt.show()
```

```python
fb=flight.pivot_table(index='month',columns='year',values='passengers')
fb
```

```python
sns.heatmap(fb,linecolor='white', linewidth=2,cmap='coolwarm',)
plt.show()
```

```python
import seaborn as sns
import scipy as sc
import matplotlib.pyplot as plt

# Load dataset
data = sns.load_dataset("penguins")

# Calculate correlation matrix
corr = data.select_dtypes(include="number").corr()


corr
```
